### VIzGradCAM Sample Usage

##### In this example, we will load a few sample images from [Pexels](https://pexels.com) and test various Keras Application [Models](https://keras.io/api/applications/) to see how it all works.

Let's load the function into this notebook!

```python
from gradcam import VizGradCAM
```

```python
import numpy as np
import tensorflow as tf
import cv2
import tensorflow.keras as keras
import matplotlib.pyplot as plt
from PIL import Image
from skimage.transform import resize
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Load Images from Pexels
!curl https://images.pexels.com/photos/1201390/pexels-photo-1201390.jpeg > monkey.jpeg
!curl https://images.pexels.com/photos/1025113/pexels-photo-1025113.jpeg > beer_glass.jpeg
!curl https://images.pexels.com/photos/1521304/pexels-photo-1521304.jpeg > cat.jpeg
```

```python
from tensorflow.keras.applications import EfficientNetB4
```

```python
test_img = img_to_array(load_img("monkey.jpeg" , target_size=(380,380)))

VizGradCAM(
    EfficientNetB4(weights="imagenet"),   #load model
    test_img
)
```

## Using MobileNetV2

```python
from tensorflow.keras.applications import MobileNetV2
```

```python
test_img = img_to_array(load_img("beer_glass.jpeg" , target_size=(224,224)))

VizGradCAM(
    MobileNetV2(weights="imagenet"), #load model
    test_img
)
```

## Using VGG16

```python
from tensorflow.keras.applications import VGG16
```

```python
test_img = img_to_array(load_img("cat.jpeg" , target_size=(224,224)))

VizGradCAM(
    VGG16(weights="imagenet"),  #load model
    test_img
)
```
